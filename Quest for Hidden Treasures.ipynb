The "Treasure Island" game is a text-based adventure where the player explores a mysterious island in search of hidden treasure.

### Introduction

The game starts with a graphical representation of a treasure map, followed by a welcome message and the player's mission to find the treasure.

### Gameplay

After the introduction, the player is prompted to make a choice at a crossroad: either to go left or right. Depending on the chosen direction, the player encounters different scenarios and challenges.

### Scenarios

- **Scenario 1 - Choosing Left**: 
    - If the player chooses to go left, they encounter a lake with an island in the middle.
    - The player must then choose whether to wait for a boat or swim across the lake.
    - If they choose to wait and arrive at the island unharmed, they are presented with three doors of different colors.
    - Depending on the door they choose, they either find the treasure, encounter danger, or choose a non-existent door.

- **Scenario 2 - Choosing Right**: 
    - If the player chooses to go right, they fall into a hole and the game ends.

### Game Over

If the player makes the wrong choices or encounters danger, they receive a "Game Over" message, and the game ends.

Overall, the player must make strategic choices to successfully navigate through the island and find the hidden treasure.

The "Mystery Mansion" game is a text-based adventure where the player explores a mysterious mansion in search of hidden treasure.

### Introduction

The game starts with a welcome message, introducing the player to the setting of the mansion and their mission to find the treasure.

### Image Display

The game attempts to display an image of the mansion using the Python Imaging Library (PIL). However, displaying images directly in the console output is not straightforward, so you might see a path to the image file printed.

### Gameplay

After the introduction, the player is prompted to choose a direction: left or right. Depending on the chosen direction, the player encounters different scenarios and challenges.

### Scenarios

- **Scenario 1 - Choosing Left or Right**: 
    - The player must choose a direction to proceed.
    
- **Scenario 2 - Left Direction (Puzzle Room)**:
    - If the player chooses to go left, they encounter a locked door and puzzles to solve.
    
- **Scenario 3 - Right Direction (Treasure Room)**:
    - If the player chooses to go right, they descend into a basement and search for the treasure.

### Game Over

If the player fails to solve a puzzle or makes the wrong choices, they encounter a "Game Over" message and the game ends.

Overall, the player must make strategic choices and solve puzzles to successfully navigate through the mansion and find the hidden treasure.


```python
from IPython.display import display
from PIL import Image

print("Welcome to Mystery Mansion!")
print("Your mission is to find the hidden treasure.")

# Path to the image file
image_path = "C:/Users/aleky/Desktop/PYTHON PROJECTS/mansion.png"

# Open and display the image inline
img = Image.open(image_path)
display(img)

# Scenario 1: Choose left or right
direction = input('You stand in front of a fork in the corridor. Which way do you want to go? "left" or "right"\n').lower()

if direction == "left":
    print("You chose to go left.")
    print("You encounter a locked door.")
    # Scenario 2 for left direction
    key = input("Do you want to search for a key to unlock the door? Type 'search' or 'ignore'\n").lower()
    if key == "search":
        print("You find a key under the doormat and unlock the door.")
        print("You enter a room filled with puzzles.")
        # Scenario 3 for left direction
        puzzle_solution = input("You must solve a riddle to proceed: 'What has keys but can't open locks?' \n").lower()
        if puzzle_solution == "keyboard":
            print("Congratulations! You solved the riddle and found the treasure. You Win!")
        else:
            print("Sorry, the answer is incorrect. You're trapped in the puzzle room. Game Over.")
    else:
        print("You decided not to search for a key. You're stuck at the locked door. Game Over.")
else:
    print("You chose to go right.")
    print("You encounter a staircase leading downwards.")
    # Scenario 2 for right direction
    staircase = input("Do you want to descend the staircase? Type 'descend' or 'turn back'\n").lower()
    if staircase == "descend":
        print("You descend the staircase and reach a dark basement.")
        print("You find a torch and light your way.")
        # Scenario 3 for right direction
        torch = input("You find yourself in a room with three chests. Which one do you open? 'gold', 'silver', or 'bronze'\n").lower()
        if torch == "gold":
            print("Congratulations! You found the treasure. You Win!")
        else:
            print("Unfortunately, you opened an empty chest. You must keep searching. Game Over.")
    else:
        print("You decided not to descend the staircase. You're stuck in the corridor. Game Over.")
```
